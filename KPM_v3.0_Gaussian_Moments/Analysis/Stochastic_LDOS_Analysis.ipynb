```python
def several_disorder():

    var = np.loadtxt("../Data/stochastic_var_Dis.dat")
    data = []

    for i in range (16):
        step = var[i*(64**3):(i + 1)*(64**3)]
        step = step/np.mean(step)
        data.append(step)

    return np.array(data)
```
