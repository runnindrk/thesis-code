```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data128 = np.loadtxt("data128.dat")[:, -1]
data256 = np.loadtxt("data256.dat")[:, -1]
data512 = np.loadtxt("data512.dat")[:, -1]
data1024 = np.loadtxt("data1024.dat")[:, -1]
data2048 = np.loadtxt("data2048.dat")[:, -1]
```

```python
plt.hist(data2048, bins = 40, label="L = 2048", density=True)
plt.legend()
plt.show()

plt.hist(data1024, bins = 40, label="L = 1024", density=True)
plt.legend()
plt.show()

plt.hist(data512, bins = 40, label="L = 512", density=True)
plt.legend()
plt.show()

plt.hist(data256, bins = 40, label="L = 256", density=True)
plt.legend()
plt.show()

plt.hist(data128, bins = 40, label="L = 128", density=True)
plt.legend()
plt.show()
```
